### Objective: To do Data Analysis, Visualization and Build Machine Learning model to Predict Customer Churn

```r
#Loading the Dataset

churn = read.csv("/kaggle/input/telecom-dataset/telco.csv")
View(churn)
```

```r
#Removing Null Values

colSums(is.na(churn))
churn = na.omit(churn)
View(churn)
```

```r
# Extract all the male senior citizens whose Payment Method is Electronic check & store the result in ‘senior_male_electronic’

senior_male_electronic = filter(churn, gender == "Male" & PaymentMethod == "Mailed check")
senior_male_electronic
```

```r
# Extract all those customers whose tenure is greater than 70 months or their Monthly charges is more than 100$ & store the result in ‘customer_total_tenure’
customer_total_tenure = filter(churn, tenure > 70 | MonthlyCharges > 100)
customer_total_tenure
```

```r
# Extract all the customers whose Contract is of two years, payment method is Mailed check & the value of Churn is ‘Yes’ & store the result in ‘two_mail_yes’
two_mail_yes = filter(churn, Contract == "Two year" & PaymentMethod == "Mailed check" & Churn == "Yes")
two_mail_yes
```

```r
# Get the count of different levels from the ‘Churn’ column

table(churn$Churn)
```

## ML Model Training and Testing

## Linear Regression:

```r
#  a. Build a simple linear model where dependent variable is ‘MonthlyCharges’ and independent variable is ‘tenure’
#   i. Divide the dataset into train and test sets in 70:30 ratio.
library(caTools)
churn$tenure = as.numeric(churn$tenure)
churn$MonthlyCharges = as.numeric(churn$MonthlyCharges)

sample.split(churn$MonthlyCharges, SplitRatio = 0.70) -> x
subset(churn, x== TRUE) -> train
subset(churn, x== FALSE) -> test

#   ii. Build the model on train set and predict the values on test set
lm(MonthlyCharges~tenure, data = train) -> linear_model
summary(linear_model)

pred = predict(linear_model, newdata = test)

#   iii. After predicting the values, find the root mean square error
library(Metrics)
rmse(pred, test$MonthlyCharges)

#   iv. Find out the error in prediction & store the result in ‘error’

cbind(Actual = test$MonthlyCharges, Predicted = pred) -> predicted_df
as.data.frame(predicted_df) -> predicted_df
cbind(predicted_df, Error = predicted_df$Actual-predicted_df$Predicted) -> predicted_df
View(predicted_df)
```

## Simple Logistic Regression:

```r
#  a. Build a simple logistic regression model where dependent variable is ‘Churn’ & independent variable is ‘MonthlyCharges’
#   i. Divide the dataset in 65:35 ratio
churn$Churn = as.factor(churn$Churn)

sample.split(churn$Churn, SplitRatio = 0.65) -> x
subset(churn, x== TRUE) -> train_1
subset(churn, x== FALSE) -> test_1

#   ii. Build the model on train set and predict the values on test set
glm(Churn~MonthlyCharges, data= train_1, family = "binomial") -> logistic_model

summary(logistic_model)

pred_1 = predict(logistic_model, newdata = test_1, type = "response")

#   iii. Build the confusion matrix and get the accuracy score
library(ROCR)
pred_train = predict(logistic_model, newdata = train_1, type = 'response')

R_pred = prediction(pred_train,train_1$Churn)

R_perf = performance(R_pred, "tpr", "fpr")
plot(R_perf)

pred_1 = ifelse(pred_1>0.4,'Yes','No')

t = table(test_1$Churn, pred_1)
t

message("The Accuracy of the Model is: ",sum(diag(t)/sum(t)))
```

## Multiple Logistic Regression

```r
#  b. Build a multiple logistic regression model where dependent variable is ‘Churn’ & independent variables are ‘tenure’ & ‘MonthlyCharges’
#   i. Divide the dataset in 80:20 ratio

sample.split(churn$Churn, SplitRatio = 0.80) -> x
subset(churn, x== TRUE) -> train_2
subset(churn, x== FALSE) -> test_2

#   ii. Build the model on train set and predict the values on test set

glm(Churn~tenure+MonthlyCharges, data= train_2, family = "binomial") -> log_model

summary(log_model)

pred_2 = predict(log_model, newdata = test_2, type = "response")

#   iii. Build the confusion matrix and get the accuracy score

pred_train1 = predict(log_model, newdata = train_2, type = 'response')

R_pred1 = prediction(pred_train1,train_2$Churn)

R_perf1 = performance(R_pred1, "tpr", "fpr")
plot(R_perf1)

pred_2 = ifelse(pred_2>0.4,'Yes','No')

t = table(test_2$Churn, pred_2)
t
message("The Accuracy of the Model is: ",sum(diag(t)/sum(t)))
```

## Decision Tree:

```r
#   a. Build a decision tree model where dependent variable is ‘Churn’ & independent variable is ‘tenure’
#     i. Divide the dataset in 80:20 ratio

sample.split(churn$Churn, SplitRatio = 0.80) -> x
subset(churn, x== TRUE) -> train_3
subset(churn, x== FALSE) -> test_3

#     ii. Build the model on train set and predict the values on test set
library(rpart)

rpart(formula = Churn~tenure,data= train_3, method= "class" ) -> dt_model

pred_3 = predict(dt_model, newdata = test_3, type = 'class')

#     iii. Build the confusion matrix and calculate the accuracy

t = table(test_3$Churn, pred_3)
t

message("The Accuracy of the Model is: ", sum(diag(t)/sum(t)))
```

## Random Forest:

```r
#   a. Build a Random Forest model where dependent variable is ‘Churn’ & independent variables are ‘tenure’ and ‘MonthlyCharges’
#     i. Divide the dataset in 70:30 ratio

sample.split(churn$Churn, SplitRatio = 0.70) -> x
subset(churn, x== TRUE) -> train_4
subset(churn, x== FALSE) -> test_4

#     ii. Build the model on train set and predict the values on test set
library(randomForest)

randomForest(formula= Churn~tenure+MonthlyCharges, data= train_4, ntree= 350)->rf_classifier

pred_4 = predict(rf_classifier, newdata = test_4, type = 'class')

#     iii. Build the confusion matrix and calculate the accuracy
t = table(test_4$Churn, pred_4)
t
message("The Accuracy of the Model is: ", sum(diag(t)/sum(t)))
```
